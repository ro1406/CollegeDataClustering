# K-Means Clustering - College Data

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
data=datasets.make_blobs(n_samples=600, n_features=4)
```

```python
data
```

```python
from sklearn.cluster import KMeans
```

## K Means - 2 Clusters

```python
model=KMeans(n_clusters=2)
```

```python
model.fit(data[0])
```

```python
model.inertia_
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
fig,axes = plt.subplots(1,2, sharey=True , figsize=(10,6))
axes[0].set_title("Original Data")
axes[1].set_title("K Means Data")

axes[0].scatter(data[0][:,0],data[0][:,1], c=data[1])
axes[1].scatter(data[0][:,0],data[0][:,1], c=model.labels_)
```

## K Means 3 Clusters

```python
model=KMeans(n_clusters=3)
model.fit(data[0])
```

```python
model.inertia_
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
fig,axes = plt.subplots(1,2, sharey=True , figsize=(10,6))
axes[0].set_title("Original Data")
axes[1].set_title("K Means Data")

axes[0].scatter(data[0][:,0],data[0][:,1], c=data[1])
axes[1].scatter(data[0][:,0],data[0][:,1], c=model.labels_)
```

## K Means 4 Clusters

```python
model=KMeans(n_clusters=4)
model.fit(data[0])
```

```python
model.inertia_
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
fig,axes = plt.subplots(1,2, sharey=True , figsize=(10,6))
axes[0].set_title("Original Data")
axes[1].set_title("K Means Data")

axes[0].scatter(data[0][:,0],data[0][:,1], c=data[1])
axes[1].scatter(data[0][:,0],data[0][:,1], c=model.labels_)
```

Inertia Keeps decreasing as number of Clusters increase, as it becomes harder to seprate the data

```python
inertiaList=[]
for k in range(2,20):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(data[0])
    inertiaList.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(12,10))
plt.plot(range(2,20),inertiaList)
plt.xlabel("K Value")
plt.ylabel("Inertia")
```

```python
inertiaList
```

Choose the point where the graph "Elbows" (Turns) is the optimal number of clusters. Stop right there!

## Clustering time

Only Private and Non-Private clusters. Also big differences in diff feats

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=2) 
```

```python
model.fit(df.drop('Private',axis=1))
```

```python
model.cluster_centers_
```

```python
model.inertia_
```

```python
df['Private']
```

Convert these strings to numerical:

```python
def change(x):
    if x=='Yes':
        return 1
    else:
        return 0
```

```python
df['Private']=df['Private'].apply(change)
```

```python
df["Private"]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df['Apps'],df['Accept'])
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df['Apps'],df['Accept'], c=df['Private'])
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df['Apps'],df['Accept'], c=model.labels_)
```

OH NO! Theyre different! Not good results :(

```python
y=df['Private']
preds=model.labels_
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Accuracy: ")
print((y==preds).mean())
print(classification_report(y,preds))
print()
print(confusion_matrix(y,preds))
```

The results are clearly very bad!

Lets try and find the Elbow value

```python
inertiaList=[]
for k in range(2,20):
    model=KMeans(n_clusters=k)
    model.fit(df)
    inertiaList.append(model.inertia_)
```

```python
plt.figure(figsize=(12,10))
plt.plot(range(2,20),inertiaList)
plt.xlabel("K Value")
plt.ylabel("Inertia")
```

```python
inertiaList
```
